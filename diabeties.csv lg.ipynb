```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *


import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("diabeties.csv")
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
xtrain ,xtest,ytrain, ytest = train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)
print(classification_report(ytest,ypred))
print(roc_auc_score(ytest,ypred))
```

```python
print(confusion_matrix(ytest,ypred))
```

```python
lg.predict_proba(xtest)
```

```python
ypredprob = lg.predict_proba(xtest)[:,1]
```

```python
ypredprob
```

```python
from sklearn.preprocessing import binarize
```

```python
ypred = binarize([ypredprob],threshold=0.20)[0]
```

```python
ypred
```

```python
print(f"Actual values : {ytest[:25].values}")
print(f"Predicted Values : {ypred[:25].astype(int)}")
```

```python
print(classification_report(ytest,ypred))
print(roc_auc_score(ytest,ypred))
```
