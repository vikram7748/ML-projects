```python
# Replace "Missing", "?", "nan" values
```

```python
re["Avg. Area House Age"].replace("missing", np.nan, inplace=True)
re["Avg. Area House Age"] = re["Avg. Area House Age"].astype("float64")
```

```python
re.dropna(how="all", subset=["Avg. Area House Age"], inplace=True)
```

```python
re["Avg. Area Number of Rooms"].replace("?", np.nan, inplace=True)
re["Avg. Area Number of Rooms"] = re["Avg. Area Number of Rooms"].astype("float64")
nr = re["Avg. Area Number of Rooms"].mean()
re["Avg. Area Number of Rooms"].fillna(nr, inplace=True)
```

```python
x = re.iloc[:, [1,2,3,4,5,7]]
x
```

```python
y = re.iloc[:, -2]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(xtrain,ytrain)
ypred = linreg.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f"MSE : {mean_squared_error(ytest, ypred)}")
print(f"RMSE : {np.sqrt(mean_squared_error(ytest, ypred))}")
print(f"R2 : {r2_score(ytest, ypred)}")
```
